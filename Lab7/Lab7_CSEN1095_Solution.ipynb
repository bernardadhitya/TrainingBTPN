## **Categorical Dependence - Challenge 2**



```
df_income = pd.read_csv('https://raw.githubusercontent.com/mervatkheir/CSEN1095-Data-Engineering/master/data/income_data.csv')
```

```
df_income.head() #inspect the dataframe
```

```
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
contigency= pd.crosstab(df_income['race'], df_income['education']) 
contigency
```

```
c,p,dof,e = chi2_contingency(contigency)
print(p)
```

Yes, there is a correlation between the race and education. This is indicated through the p value which is less than 0.05

## **Feature Engineering and Categorical Dependency - Challenge 3**

```
df_income.max()['hours-per-week'] #get the maximum of hours per week
```

```
working_hours_df = df_income.copy()

working_hours_catgorical_col = pd.cut(working_hours_df['hours-per-week'], [0,9,19,29,39,49,99],labels=['Late Hours','Moderate Late Hours','Low Late Hours','On Time','Moderate Overtime','Overtime'])

working_hours_df["working_hours_categories"] = working_hours_catgorical_col
del working_hours_df["hours-per-week"]
working_hours_df.head(10)

```

```
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
contigency= pd.crosstab(working_hours_df['working_hours_categories'], working_hours_df['gender']) 
contigency
```

```
c,p,dof,e = chi2_contingency(contigency)
print(p)
```

Since the value is 0 then definitely there is dependence between the gender and working hours.

An example of the indicator feature which we are going to come up with is: working hours indicator. If the person works more than 42 hours per week, they should be labelled as overtime workers. We might need this indication to analyze this group specifically at a later stage.
