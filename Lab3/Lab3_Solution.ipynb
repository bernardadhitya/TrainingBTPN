## **Data Tidying Exercise Solution**

```python
#read the csv
import pandas as pd 
gapminder_df = pd.read_csv("CSVs/gapminder.csv")
gapminder_df
```

```python
#fetch the columns that are related to GDP and assign them to a new dataframe
gapminder_df_gdp = gapminder_df.loc[:,'gdpPercap_1952':'gdpPercap_2007']
```

```python
# observe the dataframe
gapminder_df_gdp
```

```python
# melt the GDP-related dataframe
gapminder_df_gdp =pd.melt(gapminder_df_gdp)
gapminder_df_gdp.head()
```

```python
#rename the columns 
gapminder_df_gdp =gapminder_df_gdp.rename(columns={'value':'GDP Value'})
gapminder_df_gdp.head()
```

```python
#extract the Year from the variable column

gapminder_df_gdp['Year'] = gapminder_df_gdp['variable'].str.extract('(\d+)')
```

```python
#drop the unnecessary columns 
gapminder_df_gdp = gapminder_df_gdp.drop('variable',axis=1)
```

```python
# do the same steps for the population 

gapminder_df_pop = pd.melt(gapminder_df.loc[:,'pop_1952':'pop_2007'])
gapminder_df_pop.head()
```

```python
gapminder_df_pop['Year'] = gapminder_df_pop['variable'].str.extract('(\d+)')
```

```python
gapminder_df_pop.head()
```

```python
gapminder_df_pop = gapminder_df_pop.drop('variable',axis=1)
```

```python
gapminder_df_pop
```

```python
gapminder_df_pop = gapminder_df_pop.rename(columns={'value':'population count'})
```

```python
gapminder_df_final = pd.melt(gapminder_df.loc[:,'contient':'gdpPercap_2007'],id_vars=['continent','country'])#get the dataframe from the starting column up until gdpPercap_2007 and melt it 
```

```python
gapminder_df_final = gapminder_df_final.drop(['variable','value'],axis=1) #drop the columns var and val as they will be replaced by what has been created from the pop and gdp dfs 
```

```python
gapminder_df_final = pd.concat([gapminder_df_final,gapminder_df_pop,gapminder_df_gdp.loc[:,['GDP Value']]],axis=1)#merge the dataframes  tiegther 
```

```python
gapminder_df_final = gapminder_df_final[['continent','country','Year','population count','GDP Value']] #reorder the columns
```
